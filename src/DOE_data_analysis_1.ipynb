```python
# Load and concatenate city sheets from multiple DOE files
file_paths = [
    r"C:\Users\clint\Desktop\data\data\DOE\2019 Commercial Building Inventory - Northeast.xlsb",
    r"C:\Users\clint\Desktop\data\data\DOE\2019 Commercial Building Inventory - South Atlantic.xlsb",
    r"C:\Users\clint\Desktop\data\data\DOE\2019 Commercial Building Inventory - South Central.xlsb",
    r"C:\Users\clint\Desktop\data\data\DOE\2019 Commercial Building Inventory - West.xlsb",
    r"C:\Users\clint\Desktop\data\data\DOE\2019 Commercial Building Inventory - Midwest.xlsb"
]

# Load city sheets and concatenate them into a single DataFrame
city_dfs = [pd.read_excel(file_path, sheet_name='City', engine='pyxlsb') for file_path in file_paths]
combined_city_df = pd.concat(city_dfs, ignore_index=True)

# Rename unnamed columns
combined_city_df.columns = [col if not col.startswith('Unnamed') else f'Unnamed_{i}' for i, col in enumerate(combined_city_df.columns)]
combined_city_df
```

```python
# Keep only 'statecode', 'reported_propertytype', and 'reported_propertytype' columns in combined_city_df
combined_city_df = combined_city_df[['statecode', 'reported_propertytype', 'reported_propertytype']]
combined_city_df
```

```python
# Load the 'university-of-british-columbia-leases-2025-04-02.csv' and 'university-of-british-columbia-sales-2025-04-02.csv' files
lease_file_path = r'C:\Users\clint\Desktop\data\data\university-of-british-columbia-leases-2025-04-02.csv'
sales_file_path = r'C:\Users\clint\Desktop\data\data\university-of-british-columbia-sales-2025-04-02.csv'

lease_df = pd.read_csv(lease_file_path)
sales_df = pd.read_csv(sales_file_path)

# Isolate specific columns
lease_df = lease_df[['Property Type', 'Property Subtype', 'Property Id', 'State']]
sales_df = sales_df[['Property Type', 'Property Subtype', 'Property Id', 'State']]

# Combine the dataframes with sales_df above lease_df
combined_df = pd.concat([sales_df, lease_df], ignore_index=True)
```

```python
# Remove repeated Property Ids, keeping only the highest index row
combined_df = combined_df.loc[combined_df.groupby('Property Id').apply(lambda x: x.index.max())]
combined_df.reset_index(drop=True, inplace=True)
combined_df
```
