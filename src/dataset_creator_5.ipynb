# Enhanced CompStak Field Hierarchy for Sunburst Plot with Standard/Premium Classification
This notebook cell creates a sunburst diagram with a final outer ring indicating whether each field is standard or premium.

```python
# Import necessary libraries for visualization and data processing
import pandas as pd
import plotly.express as px
import numpy as np
from datetime import datetime

# Load the data dictionary to determine standard vs premium fields
data_dict_path = r"../Data/data_dictionary.csv"
data_dict = pd.read_csv(data_dict_path)

# Create a dictionary to map fields to their standard/premium status
field_status = {}
for index, row in data_dict.iterrows():
    field_name = row['Field']
    # Check if it's a standard field (columns 3, 4, 5 have value 1)
    is_standard = (row['Lease (Standard Data Feeds)'] == 1 or 
                  row['Sales (Standard Data Feeds)'] == 1 or 
                  row['Property (Standard Data Feeds)'] == 1)
    # Check if it's a premium field (columns 6, 7 have value 1)
    is_premium = (row['Tenant (Premium Add Ons)'] == 1 or 
                 row['Property (Premium Add Ons)'] == 1)
                 
    if is_standard:
        field_status[field_name] = 'Standard'
    elif is_premium:
        field_status[field_name] = 'Premium'
    else:
        field_status[field_name] = 'Unknown'  # For any field not found in the criteria

# Manually set "Tenant Revenue" as premium since we know it is a premium field
field_status['Tenant Revenue'] = 'Premium'

# Convert the sunburst_df list into a pandas DataFrame
sunburst_data = pd.DataFrame(sunburst_df, columns=['Level1', 'Level2', 'Level3', 'Field'])

# Add the standard/premium classification as a new column
sunburst_data['Status'] = sunburst_data['Field'].map(lambda x: field_status.get(x, 'Unknown'))

# Create a path column required for the sunburst plot
sunburst_data['path'] = sunburst_data.apply(
    lambda row: [row['Level1'], row['Level2'], row['Level3'], row['Field'], row['Status']], 
    axis=1
)

# Create the sunburst plot with the Status as the outermost ring
fig = px.sunburst(
    sunburst_data, 
    path=['Level1', 'Level2', 'Level3', 'Field', 'Status'],
    title='CompStak Data Field Hierarchy with Standard/Premium Classification',
    width=1000,
    height=800,
    color='Status',
    color_discrete_map={'Standard':'#2E86C1', 'Premium':'#E74C3C', 'Unknown':'#7F8C8D'}
)

# Update layout for better visualization
fig.update_layout(
    margin=dict(t=50, l=0, r=0, b=0),
    font=dict(size=12)
)

# Generate a timestamp for the filename
timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
output_filename = f'../Images/standard_premium_sunburst_{timestamp}.html'

# Show the plot
fig.show()

# Save the interactive plot as HTML file
fig.write_html(output_filename)

# Also save as static image
static_filename = f'../Images/standard_premium_sunburst_{timestamp}.jpg'
fig.write_image(static_filename)

print(f"Interactive sunburst diagram saved to {output_filename}")
print(f"Static image saved to {static_filename}")

# Count standard vs premium fields
status_counts = sunburst_data['Status'].value_counts()
print("\nField classification counts:")
print(status_counts)
print(f"\nPercentage of standard fields: {status_counts.get('Standard', 0)/len(sunburst_data)*100:.1f}%")
print(f"Percentage of premium fields: {status_counts.get('Premium', 0)/len(sunburst_data)*100:.1f}%")
```

## Field Count Analysis for CompStak Data

```python
# Count the number of fields in each level of the hierarchy
total_fields = len(sunburst_data)
print(f"Total number of fields in the sunburst diagram: {total_fields}")

# Count by Level 1 (Top-level categories)
level1_counts = sunburst_data.groupby('Level1').size().reset_index(name='Field Count')
level1_counts = level1_counts.sort_values('Field Count', ascending=False)

# Display counts for Level 1
print("\nField counts by top-level category:")
print(level1_counts.to_string(index=False))

# Display percentage of total for each Level 1 category
level1_counts['Percentage'] = (level1_counts['Field Count'] / total_fields * 100).round(2)
print("\nPercentage distribution of fields:")
print(level1_counts[['Level1', 'Percentage']].to_string(index=False))

# Count by Level 1 and Level 2 (nested categories)
level2_counts = sunburst_data.groupby(['Level1', 'Level2']).size().reset_index(name='Field Count')
level2_counts = level2_counts.sort_values(['Level1', 'Field Count'], ascending=[True, False])

print("\nDetailed field counts by category and subcategory:")
for level1 in sunburst_data['Level1'].unique():
    print(f"\n{level1}:")
    subset = level2_counts[level2_counts['Level1'] == level1]
    for _, row in subset.iterrows():
        print(f"  {row['Level2']}: {row['Field Count']} fields")
        
        # Get further breakdown to Level 3
        level3_counts = sunburst_data[
            (sunburst_data['Level1'] == level1) & 
            (sunburst_data['Level2'] == row['Level2'])
        ].groupby('Level3').size().reset_index(name='count')
        
        for _, l3_row in level3_counts.iterrows():
            print(f"    - {l3_row['Level3']}: {l3_row['count']} fields")

# Create a bar chart of the top-level categories
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.bar(level1_counts['Level1'], level1_counts['Field Count'], color='skyblue')
plt.xlabel('Category')
plt.ylabel('Number of Fields')
plt.title('CompStak Data Fields by Category')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Save the chart
from datetime import datetime
timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
plt.savefig(f'../Images/field_count_by_category_{timestamp}.png')
plt.show()
```
