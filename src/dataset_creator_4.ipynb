```python
import pandas as pd

# Load the CSV file
file_path = r"C:\Users\clint\Desktop\compstak-analysis\Data\data_dictionary.csv"
data = pd.read_csv(file_path)

# Display the first few rows of the data
data
```

```python
# Create proper column names by combining row 0 and row 1
# Extract the first two rows for headers
header_row0 = data.iloc[0]
header_row1 = data.iloc[1]

# Create new column names by combining row 0 and row 1 where both exist
new_column_names = []
for col in data.columns:
    if pd.notna(header_row0[col]) and pd.notna(header_row1[col]):
        new_column_names.append(f"{header_row1[col]} ({header_row0[col]})")
    elif pd.notna(header_row1[col]):
        new_column_names.append(header_row1[col])
    else:
        new_column_names.append(col)

# Create a new dataframe with proper data (skip the first two rows)
formatted_data = data.iloc[2:].reset_index(drop=True)
formatted_data.columns = new_column_names

# Display the reformatted dataframe
formatted_data
```

```python
# Add a new cell to convert X to 1 and missing values to 0
binary_data = formatted_data.copy()

# Replace 'X' with 1 and NaN/empty with 0
binary_data = binary_data.applymap(lambda x: 1 if x == 'X' else (0 if pd.isna(x) or x == '' else x))
binary_data = binary_data.fillna(0)
# Display the binary representation
binary_data
```

```python
# Drop 'Data Type' and 'Example Value' columns
columns_to_drop = [col for col in binary_data.columns if 'Data Type' in col or 'Example Value' in col]
filtered_data = binary_data.drop(columns=columns_to_drop)

# Display the filtered dataframe
print(f"Dropped {len(columns_to_drop)} columns: {columns_to_drop}")
filtered_data
```
