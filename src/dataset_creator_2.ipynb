```python
import pandas as pd

# Load the Excel file
file_path = r"C:\Users\clint\Desktop\data\data\CompStak Data Dictionary (3).xlsx"
data = pd.ExcelFile(file_path)

data
```

```python
# Read and display the single sheet in the Excel file
sheet_name = data.sheet_names[0]
print(f"Sheet name: {sheet_name}")

# Load the data into a DataFrame
df = pd.read_excel(data, sheet_name=sheet_name)

# Display the first several rows of the data
print(f"\nPreview of the data (first 10 rows):")
display(df.head(10))

# Display basic information about the dataframe
print("\nDataframe info:")
df.info()

print("\nDataframe shape (rows, columns):", df.shape)

# Display summary statistics
print("\nSummary statistics for numerical columns:")
display(df.describe())
```

```python
# Save the DataFrame as a CSV file
output_path = r"C:\Users\clint\Desktop\compstak-analysis\Data\data_dictionary.csv"

# Save the DataFrame to CSV
df.to_csv(output_path, index=False)

print(f"DataFrame successfully saved to: {output_path}")
```
