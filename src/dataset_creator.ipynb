# Interactive Column Selection Tool

This cell runs a tool that will help you go through each column in the dataset and decide whether to keep or delete it.
After each decision, you'll be asked to provide reasoning. You can also go back to correct any previous decisions.

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import pandas as pd

# Initialize dictionaries to store decisions and reasons
column_decisions = {}
column_reasons = {}

# Get all column names
columns = list(data.columns)
total_columns = len(columns)
current_index = [0]  # Use list to make it mutable in nested functions

# Create widgets
col_name_display = widgets.HTML()
col_info_display = widgets.HTML()
progress_display = widgets.HTML()
keep_button = widgets.Button(description="Keep", button_style="success")
delete_button = widgets.Button(description="Delete", button_style="danger")
back_button = widgets.Button(description="Go Back", button_style="info")
reason_input = widgets.Textarea(description="Reason:", placeholder="Why did you make this decision?", layout=widgets.Layout(width="80%", height="100px"))
submit_reason_button = widgets.Button(description="Submit & Continue", button_style="primary")
reason_output = widgets.Output()
final_output = widgets.Output()

# Function to display column info
def display_column_info(index):
    if index < 0:
        # Cannot go back past the first column
        current_index[0] = 0
        display_column_info(0)
        return
        
    if index < total_columns:
        col_name = columns[index]
        col_name_display.value = f"<h3>Column {index+1}/{total_columns}: <span style='color:blue'>{col_name}</span></h3>"
        
        # Collect information about the column
        dtype = data[col_name].dtype
        non_null_count = data[col_name].count()
        null_percentage = round((len(data) - non_null_count) / len(data) * 100, 2)
        unique_values = data[col_name].nunique()
        
        # Display sample values
        sample_values = data[col_name].head().tolist()
        sample_str = ', '.join([str(val) for val in sample_values])
        if len(sample_str) > 300:  # Truncate if too long
            sample_str = sample_str[:300] + '...'
            
        # Create info HTML
        col_info = f"""
        <b>Data Type:</b> {dtype}<br>
        <b>Non-null Count:</b> {non_null_count} / {len(data)} ({100-null_percentage}%)<br>
        <b>Null Percentage:</b> {null_percentage}%<br>
        <b>Unique Values:</b> {unique_values}<br>
        <b>Sample Values:</b> {sample_str}<br>
        """
        col_info_display.value = col_info
        progress_display.value = f"<b>Progress:</b> {index+1}/{total_columns} columns reviewed"
        
        # Show action buttons
        button_box = widgets.HBox([keep_button, delete_button])
        if index > 0:
            button_box = widgets.HBox([keep_button, delete_button, back_button])
            
        display(col_name_display, col_info_display, progress_display)
        display(button_box)
    else:
        # All columns processed, show final results
        show_final_results()

def collect_reason(decision, col_name):
    """Collect reasoning for a decision"""
    clear_output()
    # Show what decision was made
    decision_text = "keep" if decision == "keep" else "delete"
    decision_html = widgets.HTML(f"<h3>You decided to <span style='color:{'green' if decision == 'keep' else 'red'}'>{decision_text}</span> the column <span style='color:blue'>{col_name}</span></h3>")
    
    display(decision_html)
    display(reason_input)
    display(submit_reason_button)

# Function to handle keep button click
def on_keep_click(b):
    col_name = columns[current_index[0]]
    column_decisions[col_name] = 'keep'
    collect_reason("keep", col_name)

# Function to handle delete button click
def on_delete_click(b):
    col_name = columns[current_index[0]]
    column_decisions[col_name] = 'delete'
    collect_reason("delete", col_name)

# Function to handle back button click
def on_back_click(b):
    if current_index[0] > 0:
        current_index[0] -= 1
        # Remove previous decision and reason if they exist
        col_name = columns[current_index[0]]
        if col_name in column_decisions:
            del column_decisions[col_name]
        if col_name in column_reasons:
            del column_reasons[col_name]
    clear_output()
    display_column_info(current_index[0])

# Function to handle reason submission
def on_submit_reason(b):
    col_name = columns[current_index[0]]
    reason = reason_input.value
    
    # Store the reason
    column_reasons[col_name] = reason
    
    # Clear the reason input for the next column
    reason_input.value = ""
    
    # Move to next column
    current_index[0] += 1
    clear_output()
    
    if current_index[0] < total_columns:
        display_column_info(current_index[0])
    else:
        show_final_results()

# Function to show final results
def show_final_results():
    with final_output:
        clear_output()
        kept_cols = [col for col, decision in column_decisions.items() if decision == 'keep']
        deleted_cols = [col for col, decision in column_decisions.items() if decision == 'delete']
        
        print("\n=== COLUMN SELECTION SUMMARY ===")
        print(f"\nKept {len(kept_cols)} columns, Deleted {len(deleted_cols)} columns\n")
        
        print("=== KEPT COLUMNS AND REASONS ===")
        for col in kept_cols:
            reason = column_reasons.get(col, "No reason provided")
            print(f"- {col}: {reason}")
            
        print("\n=== DELETED COLUMNS AND REASONS ===")
        for col in deleted_cols:
            reason = column_reasons.get(col, "No reason provided")
            print(f"- {col}: {reason}")
        
        # Create filtered dataframe
        print("\n=== FILTERED DATAFRAME ===")
        filtered_df = data[kept_cols]
        print(f"\nShape: {filtered_df.shape}")
        display(filtered_df.head())
        
        # Export variable for later use
        print("\nThe filtered DataFrame is available as 'filtered_data'")
        print("You can also access the decisions and reasons as 'column_decisions' and 'column_reasons'")

    # Create the filtered dataframe as a global variable
    global filtered_data
    filtered_data = data[[col for col, decision in column_decisions.items() if decision == 'keep']]
    
    display(final_output)

# Set button callbacks
keep_button.on_click(on_keep_click)
delete_button.on_click(on_delete_click)
back_button.on_click(on_back_click)
submit_reason_button.on_click(on_submit_reason)

# Start the process
display_column_info(0)
```
